```python
with open("news.txt", "r", encoding = "UTF-8") as f:
    data = f.read()
```

```python
data
```

```python
data = list(data)
data = list(filter(str.isalpha, data))
```

```python
data
```

```python
data = list(map(str.lower, data))
```

```python
data[0]
```

```python
from functools import reduce

data = reduce(
        lambda r, e:
            r.update({e : r.get(e, 0) + 1})or r,
        data,
        {}
)

data
```

```python
data = sorted(data.items(), reverse = True,  key = lambda x : x[1])
```
