```python
def sender(data, target):
    next(target)
    for e in data:
        b_sent = target.send(e)
        if b_sent == e:
            print("data sent")
        elif b_sent == None:
            pass
    target.close()
    
def receiver(target):
    print("receiver executed")
    next(target)
    ex_data = None
    try:
        while True:
            data = (yield ex_data)
            received = target.send(data)
            ex_data = data
            
            if received == data:
                print("data from receiver to printer")
            elif received == None:
                pass
    except GeneratorExit:
        print("receiver shutdown")
        target.close()
        
def printer():
    print("printer excecuted")
    ex_data = None
    try:
        while True:
            data = (yield ex_data)
            print("data : {data}".format(data = data), end = '   ')
            ex_data = data           
    except GeneratorExit:
        print("print shutdown")
        
data = [k ** 2 for k in range(1, 34)]
sender(data, receiver(printer()))
```
